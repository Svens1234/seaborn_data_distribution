```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
dots=sns.load_dataset('dots')
```

```python
dots.head()
```

```python
sns.distplot(dots['firing_rate'])
```

```python
sns.distplot(dots['firing_rate'], kde=False)
```

```python
sns.distplot(dots['firing_rate'], kde=False, bins=30)
```

```python
sns.distplot(dots['firing_rate'], kde=False, bins=100)
```

```python
sns.jointplot(x='coherence', y='firing_rate', data=dots)
```

```python
tips= sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.distplot(tips['total_bill'], kde=False, bins=20)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde')
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips, hue='sex')
```

```python
sns.pairplot(tips, hue='sex', palette='husl')
```
